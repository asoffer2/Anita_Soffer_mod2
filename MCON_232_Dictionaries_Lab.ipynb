```python
import copy
my_dict2 = {"name" : "LAS",
            "age" : "street",
            "pet-type" : 100

}

my_dict3 = my_dict | my_dict2 # combines 2
print(my_dict3)
my_dict3 |= my_dict2
print(my_dict3) # wont add in non unique keys, over-writes values if same key.

# for everything: deep copies. if u just do a copy, only copies the surface
my_dict5 = copy.deepcopy(my_dict3)
print(my_dict5)

my_dict5 = my_dict3.copy() # uses less power
print(my_dict5)
```

```python
#leetcode:

def leetCode(s):
  letters_dict = {}
  for char in s:
    count = 0
    letter = char
    for ch in s:
      if letter == ch:
        count += 1
    letters_dict[f"{char}"] = count
  non_repeating_letter = None
  for k, v in letters_dict.items():
    if v == 1:
      non_repeating_letter = k
      break
  if non_repeating_letter is not None:
    index = s.find(non_repeating_letter)
  else:
    index = -1
  return index


print(leetCode("heeeyyy"))
print(leetCode("hhheeeyyy"))
print(leetCode("hhelllo"))
```
